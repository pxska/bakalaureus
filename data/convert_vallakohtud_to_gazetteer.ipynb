```python
import csv
import os
import re

from estnltk.converters import json_to_text
```

```python
data_csv = []
with open('vallakohtud.csv', 'r', encoding='UTF-8') as file:
    reader = csv.reader(file, delimiter='|')
    for row in reader:
        data_csv.append(row)
```

```python
leave_out = []

with open('id_left_out.txt', 'r', encoding='UTF-8') as f:
    ids = f.readlines()

for i in ids:
    leave_out.append(i.rstrip('\n'))
```

```python
pattern_person = '<person .*?>(.*?)<\/person>'
pattern_location = '<location .*?>(.*?)<\/location>'
pattern_unknown = '<unknown .*?>(.*?)<\/unknown>'
```

```python
set_of_names = set()
counter = 1
for protocol in data_csv:
    if protocol[1] == 'text' or str(protocol[0]) in leave_out:
        print(f' {counter}. Jätan välja protokolli id-ga {protocol[0]}.')
        counter += 1
        continue
    else:
        match = re.search(pattern_person, protocol[1])
        if match:
            location_match = re.search(pattern_location, match.group(1).strip())
            unknown_match = re.search(pattern_unknown, match.group(1).strip())
            if location_match:
                set_of_names.add(location_match.group(1).strip() + '\t' + 'LOC')
                continue
            if unknown_match:
                set_of_names.add(unknown_match.group(1).strip() + '\t' + 'MISC')
                continue
            set_of_names.add(match.group(1).strip() + '\t' + 'PER')
```

```python
names_in_protocols = set()
```

```python
for file in os.listdir(os.path.join('..', 'data', 'vallakohtufailid-json-flattened')):
    with open(os.path.join('..', 'data', 'vallakohtufailid-json-flattened', file), 'r', encoding='UTF-8') as json_file:
        text = json_to_text(json_file.read())
        for span in text.gold_wordner:
            if span.nertag != 'O':
                names_in_protocols.add(span.text)
```

```python
new_set_of_names = set_of_names.difference(names_in_protocols)
```

```python
with open('uus_vk_gazetteer.txt', 'w+', encoding='UTF-8') as file:
    for name in new_set_of_names:
        file.write(str(name)+'\n')
```
